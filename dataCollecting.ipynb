```python
#Cria variáveis para abertura dos arquivos TXT onde estão os elementos das páginas que contém as tags a serem manipuladas
#São dois arquivos por ano por causa da paginação do Discogs
html_lista1_2003_file = open('html_listitems_elements/rs_2003_elemento_listitems_lista1.txt','r')
html_lista2_2003_file = open('html_listitems_elements/rs_2003_elemento_listitems_lista2.txt','r')
html_lista1_2020_file = open('html_listitems_elements/rs_2020_elemento_listitems_lista1.txt','r')
html_lista2_2020_file = open('html_listitems_elements/rs_2020_elemento_listitems_lista2.txt','r')
```

```python
#Cria variáveis para leitura dos arquivos TXT abertos na célula anterior
html_lista1_2003 = html_lista1_2003_file.read()
html_lista2_2003 = html_lista2_2003_file.read()
html_listaFinal_2003 = html_lista1_2003 + html_lista2_2003

html_lista1_2020 = html_lista1_2020_file.read()
html_lista2_2020 = html_lista2_2020_file.read()
html_listaFinal_2020 = html_lista1_2020 + html_lista2_2020
```

```python
#Cria variáveis do módulo BeautifulSoup e atribui as variáveis dos arquivos carrregados sob o método html.parser (analisador de html)
soup2003 = BeautifulSoup(html_listaFinal_2003,'html.parser')

soup2020 = BeautifulSoup(html_listaFinal_2020,'html.parser')
```

```python
#Cria uma variável de lista vazia que armazenará os IDs de cada master/release da lista da Rolling Stones
masterID_lista2003 = []

masterID_lista2020 = []

#Executa iteração com a variável do BeautifulSoup para encontrar as tags conforme parâmetros:
  #Para cada tag 'li', traga os elementos da class 'listitem body_row row_with_actions shortcut_navigable', partindo do 17º caracter da tag 'id' e inclua na variável de lista vazia
  #Importante: é necessário analisar o html da página para identificar o padrão de onde estão os dados desejados
for i in soup2003.find_all('li', attrs={'class':'listitem body_row row_with_actions shortcut_navigable'}):
    masterID_lista2003.append(i['id'][16:])

for i in soup2020.find_all('li', attrs={'class':'listitem body_row row_with_actions shortcut_navigable'}):
    masterID_lista2020.append(i['id'][16:])
```

```python
#Cria funções para coleta de dados dos álbuns na API do Discogs

dClient = discogs_client.Client('collectingDiscogs', user_token = '')

def get_artistName(ID, MasterReleaseType):
    if MasterReleaseType == 'M':
        masterRelease = dClient.master(ID)
        mainRelease = masterRelease.main_release
        artistStr = str(mainRelease.artists)
    elif MasterReleaseType == 'R':
        release = dClient.release(ID)
        artistStr = str(release.artists)
    if artistStr.find('"')>0:
        slcStart_artist = artistStr.find('"')+1
        slcEnd_artist = artistStr.find('"',slcStart_artist)
        artistName = artistStr[slcStart_artist:slcEnd_artist]
    elif artistStr.find("'")>0:
        slcStart_artist = artistStr.find("'")+1
        slcEnd_artist = artistStr.find("'",slcStart_artist)
        artistName = artistStr[slcStart_artist:slcEnd_artist]
    return artistName


def get_labelName(ID, MasterReleaseType):
    if MasterReleaseType == 'M':
        masterRelease = dClient.master(ID)
        mainRelease = masterRelease.main_release
        labelStr = str(mainRelease.labels)
    elif MasterReleaseType == 'R':
        release = dClient.release(ID)
        labelStr = str(release.labels)
    if labelStr.find('"')>0:
        slcStart_label = labelStr.find('"')+1
        slcEnd_label = labelStr.find('"', slcStart_label)
        labelName = labelStr[slcStart_label:slcEnd_label]
    elif labelStr.find("'")>0:
        slcStart_label = labelStr.find("'")+1
        slcEnd_label = labelStr.find("'", slcStart_label)
        labelName = labelStr[slcStart_label:slcEnd_label]
    return labelName

def get_masterTitle(ID, MasterReleaseType):
    if MasterReleaseType == 'M':
        masterRelease = dClient.master(ID)
        title = masterRelease.title
    elif MasterReleaseType == 'R':
        release = dClient.release(ID)
        title = release.title
    return title

def get_masterGenre(ID, MasterReleaseType):
    if MasterReleaseType == 'M':
        masterRelease = dClient.master(ID)
        genres = masterRelease.genres
    elif MasterReleaseType == 'R':
        release = dClient.release(ID)
        genres = release.genres
    return genres

def get_masterStyle(ID, MasterReleaseType):
    if MasterReleaseType == 'M':
        masterRelease = dClient.master(ID)
        styles = masterRelease.styles
    elif MasterReleaseType == 'R':
        release = dClient.release(ID)
        styles = release.styles
    return styles

def get_masterYear(ID, MasterReleaseType):
    if MasterReleaseType == 'M':
        masterRelease = dClient.master(ID)
        mainRelease = masterRelease.main_release
        year = mainRelease.year
    elif MasterReleaseType == 'R':
        release = dClient.release(ID)
        year = release.year
    return year
```

```python
#Cria a lista que receberá os dados coletados
mastersAtributes = []
```

```python
#Cria o dicionário com os dados coletados da lista de 2003 (loop cabe aprimoramento)

rank = 1

for i in masterID_lista2003:
    ID = i[2:]
    MasterReleaseType = i[0]
    try:
        row = {
            'year_list': 2003,
            'rank': rank,
            'master_release_id': i,
            'artist_name': get_artistName(ID, MasterReleaseType),
            'label_name': get_labelName(ID, MasterReleaseType),
            'genre': get_masterGenre(ID, MasterReleaseType),
            'style': get_masterStyle(ID, MasterReleaseType),
            'title': get_masterTitle(ID, MasterReleaseType),
            'year' : get_masterYear(ID, MasterReleaseType)
        }
    except:
        pass
 
    mastersAtributes.append(row)
    print(row)
    rank += 1
```

```python
#Cria o dicionário com os dados coletados da lista de 2020 (loop cabe aprimoramento)

rank = 1

for i in masterID_lista2020:
    ID = i[2:]
    MasterReleaseType = i[0]
    try:
        row = {
            'year_list': 2020,
            'rank': rank,
            'master_release_id': i,
            'artist_name': get_artistName(ID, MasterReleaseType),
            'label_name': get_labelName(ID, MasterReleaseType),
            'genre': get_masterGenre(ID, MasterReleaseType),
            'style': get_masterStyle(ID, MasterReleaseType),
            'title': get_masterTitle(ID, MasterReleaseType),
            'year' : get_masterYear(ID, MasterReleaseType)
        }
    except:
        pass
 
    mastersAtributes.append(row)
    print(row)
    rank += 1
```

```python
#Data cleaning

for i in mastersAtributes:
  if i['artist_name'].count('(')>0:
    i['artist_name'] = i['artist_name'][:(i['artist_name']).find('(')-1]
  if i['label_name'].count('(')>0:
    i['label_name'] = i['label_name'][:(i['label_name']).find('(')-1]
```

```python
#Cria o dataset json com os dados coletados e limpos

jsonFile = open('rs_bestAlbums_dataset.json', 'w')
json.dump(mastersAtributes, jsonFile)
jsonFile.close()
```
